```python
import cv2
import numpy as np
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cmath


weight_1layer_alexnet = np.load("weight_1layer_alexnet.npy")
target_layer_weight = weight_1layer_alexnet[0]
# weight_1layer_alexnet[1]


#(11,11,3,96)


num_filta = len(target_layer_weight[0,0,0,:])  #96
num_channel = len(target_layer_weight[0,0,:,0])  #3

fimage = np.zeros((3,96,11,11),dtype='complex128') 
max_abs=np.zeros(288).reshape((3,96))
max_freq=np.zeros(288).reshape((3,96))

x=np.zeros(288).reshape((3,96))
y=np.zeros(288).reshape((3,96))

for channel in range(num_channel):
    for filta in range(num_filta):
        data = target_layer_weight[:,:,channel,filta]
        fimage[channel,filta,:,:]= np.fft.fft2(data)
        fimage_abs = np.abs(fimage[channel,filta,:,:])
        #fimage_abs = math.sqrt(fimage[channel,filta,:,:].real**2 *fimage[channel,filta,:,:].imag**2)
        maxindex=np.argmax(fimage_abs)
        maxrow = int(maxindex/11) 
        maxcol = int(maxindex%11)
        #print(maxindex)
        temp_polar = cmath.polar(fimage[channel,filta,maxrow,maxcol])
        max_abs[channel][filta] = temp_polar[0]
        max_freq[channel][filta] = temp_polar[1]
        x[channel][filta] = max_abs[channel][filta]*math.cos(max_freq[channel][filta])
        y[channel][filta] = max_abs[channel][filta]*math.sin(max_freq[channel][filta])
        #print(max_abs[channel][filta])
        
        
        
        #print(max_freq[channel][filta])
        


    
        #heatmap = ax.pcolor(data, cmap=plt.cm.binary) 
        #ax.set_xticklabels(row_labels, minor=False)
        #ax.set_yticklabels(column_labels, minor=False)
#plt.show()


fig = plt.figure()
plt.polar(max_freq[0,:],max_abs[0,:],'ro',)
plt.savefig("alexnet_1st_layer_filter_channel0.png")
fig = plt.figure()
plt.polar(max_freq[1,:],max_abs[1,:],'ro')
plt.savefig("alexnet_1st_layer_filter_channel1.png")
fig = plt.figure()
plt.polar(max_freq[2,:],max_abs[2,:],'ro')
plt.savefig("alexnet_1st_layer_filter_channel2.png")



fig = plt.figure()
plt.polar(max_freq[0,:],max_abs[0,:],'ro',)
plt.polar(max_freq[1,:],max_abs[1,:],'ro')
plt.polar(max_freq[2,:],max_abs[2,:],'ro')
plt.savefig("alexnet_1st_layer_filter_sum_of_3_channels.png")


fig = plt.figure()
plt.polar(max_freq[0,:],max_abs[0,:],'ro',)
plt.polar(max_freq[1,:],max_abs[1,:],'go')
plt.polar(max_freq[2,:],max_abs[2,:],'bo')
plt.savefig("alexnet_1st_layer_filter_sum_of_3_channels_color_changed.png")




plt.show()


#print(max_abs.shape)
#print(fimage_abs)
'''
fimage_polar = cmath.polar(fimage)
plt.plot(fimage_abs)
print(fimage_abs)
#print(fimage)
print(fimage_abs.shape)



 
#image = cv2.imread("fujji.png", cv2.IMREAD_GRAYSCALE)
#cv2.imshow("original",image)
#cv2.waitKey(0)
#cv2.destroyWindow()
""""
fimage = np.fft.fft2(image)
print(fimage)
print(fimage.shape)
"""
'''
```
