```python
#最大値の半値をこすような面積を求める
def cut_percent(z,p=0.5):
    #pは残る割合（0<p<1）指定しない場合は0.5
    z_max_half = (1-p)*z.max()
    z_cutted = z>z_max_half
    z_cutted = 1*z_cutted
    return z_cutted

def normalization(z):
    #z>=0を仮定
    out = z/z.max()
    return out

total_z_cutted = np.zeros(z.shape)


percent = np.arange(0.1,1.1,0.1)
plt.figure(figsize=(15,6))

for ii in percent:
    for filta_th in arange(96):
        weight081 = target_layer_weight[:,:,0,filta_th]
        ffted_weight081= np.fft.fft2(weight081,s=[65,65])
        ffted_shifted_weight081= np.fft.fftshift(ffted_weight081)
        abs_ffted_weight081 = abs(ffted_shifted_weight081)
        z = abs_ffted_weight081
        z_cutted = cut_percent(z,ii)
        total_z_cutted += z_cutted
    
#normalizated_z = normalization(total_z_cutted)
    
    ii *=10
    plt.subplot(2,5,ii)
    plt.subplots_adjust(wspace=0.3,hspace=0.2)
    plt.imshow(total_z_cutted)
    plt.colorbar()
    plt.title(str(round(ii/10,1)))
    total_z_cutted = np.zeros(z.shape)
    
plt.savefig("total_z_cutted.png")
plt.show()


```
